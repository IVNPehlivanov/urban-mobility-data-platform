```python

from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
import ConnectionConfigKaloyan as cc
cc.setupEnvironment()
```

```python
spark = cc.startLocalCluster("DIM_WEATHER",1)
spark.getActiveSession()
```

```python
# Define schema
weather_schema = StructType([
    StructField("weather_id", IntegerType(), False),  # Primary Key
    StructField("weather_condition", StringType(), True)  # Weather Description
])

# Define the known weather types
weather_data = [
    Row(weather_id=1, weather_condition="Unpleasant"),
    Row(weather_id=2, weather_condition="Pleasant"),
    Row(weather_id=3, weather_condition="Neutral"),
    Row(weather_id=0, weather_condition="Unknown")
]
```

```python
# Create DataFrame
df_dim_weather = spark.createDataFrame(weather_data, schema=weather_schema)
```

```python
# Save as Delta table
#df_dim_weather.write.format("delta").mode("overwrite").saveAsTable("dimWeather")

#parquet files
df_dim_weather.repartition(1).write.format("parquet").mode("overwrite").saveAsTable("dimWeather_pq")
```

```python
# Show result
df_dim_weather.show()
```

```python
spark.stop()
```
